```python
import random
# Assuming new_balanced_set is a dictionary containing the categories and their samples.

# Function to create a new dataset with the specified percentage for each category
def create_custom_percentage_dataset(dataset, custom_percentage_category):
    new_dataset = {}
    for category, samples in dataset.items():
        # Set the percentage based on the category
        percentage = 0.15 if category == custom_percentage_category else 0.08
        # Calculate the length of samples based on the percentage
        new_length = int(len(samples) * percentage)
        # Shuffle the samples to randomly select the desired percentage
        random.shuffle(samples)
        # Add the selected percentage of samples to the new dataset
        new_dataset[category] = samples[:new_length]
        # Remove the samples from the original dataset
        dataset[category] = samples[new_length:]
    return new_dataset

# Call the function to create the new dataset with custom percentages
custom_percentage_category = "Environmental Protection"
new_dataset_custom_percentage = create_custom_percentage_dataset(new_balanced_set, custom_percentage_category)

# Print the result
for k, v in new_dataset_custom_percentage.items():
    print(k, len(v))
# Now the new_balanced_set will be updated without the samples that were added to new_dataset_custom_percentage
```
